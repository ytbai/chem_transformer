```
import torch
import numpy as np
import pandas as pd
import os

from data_factory import *
from model_factory import *
```

```
## This notebook is for testing the main Transformer model
## that predicts log solubility.

# Create instance of model
model = ChemTransformer().cuda()
model_factory = ModelFactory(model)

# Load saved weights
model_factory.load()

metric = nn.MSELoss()
```

```
# Load test data set
test_dataset = DelaneyDataset(mode = "test")

```

```
# RMSE test result for model
np.sqrt(test(model_factory, test_dataset, metric))
```

```
# RMSE test result for ESOL
np.sqrt(test("esol", test_dataset, metric))
```

```
# RMSE test result for mean model
train_dataset = DelaneyDataset(mode = "train")
target_mean = train_dataset.set_target("y_true").get_target_mean()
np.sqrt(test(target_mean, test_dataset, metric))
```
