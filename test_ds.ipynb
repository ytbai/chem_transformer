```
# Create an instance of model_ds
model_ds = ModelDS(target_name)
# Load saved weights
model_ds.model_factory.load()

metric = nn.MSELoss()

# Create an instance of "model", which is the main model pre-trained to
# predict log solubility. 
model = ChemTransformer().cuda()
model_factory = ModelFactory(model)
# Load saved weights
model_factory.load()
```

```
# Load test dataset
test_dataset = DelaneyDataset(mode = "test")
```

```
# RMSE test result for model_ds
np.sqrt(test_ds(model_ds, model_factory, test_dataset, metric))
```
